```python
from ipyleaflet import Map, basemaps, Marker

# # Map centred on (60 degrees latitude et -2.2 degrees longitude)
# # Latitude, longitude
map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python

coordinates = ds2[['latitude', 'longitude']].values
populations = ds2['level5'].values
test = True
# Adding Markers to the map with Popup
for (lat, lon), pop in zip(coordinates, populations):
    if lat != '0' and lon != '0':
        marker = Marker(location=(lat, lon))
        map.add_layer(marker)
    else:
        continue
map
```
